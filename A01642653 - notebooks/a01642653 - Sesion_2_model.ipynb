## Implement preprocessing steps

### Subtask:
Based on the variable types identified earlier, apply appropriate preprocessing steps such as one-hot encoding for categorical features and scaling for numerical features. This will be done using ColumnTransformer to create a robust preprocessing pipeline.


**Reasoning**:
Define numerical and categorical features and create a ColumnTransformer for preprocessing.



```python
numerical_features = model_data_train.select_dtypes(include=np.number).columns.tolist()
categorical_features = categorical_variables

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])
```

## Define a pipeline for each model

### Subtask:
Create a scikit-learn pipeline for each selected model, combining the defined preprocessor and the model itself.


**Reasoning**:
Create scikit-learn pipelines for each selected model, combining the preprocessor and the model.



```python
models = [('RandomForest', RandomForestClassifier(random_state=42)),
          ('GradientBoosting', GradientBoostingClassifier(random_state=42)),
          ('LogisticRegression', LogisticRegression(solver='liblinear', random_state=42))]

pipelines = {}
for name, model in models:
    pipelines[name] = Pipeline(steps=[('preprocessor', preprocessor),
                                      ('classifier', model)])
```

## Set up hyperparameter tuning

### Subtask:
Define hyperparameter grids for each model to tune them effectively for better performance on the imbalanced dataset.


**Reasoning**:
Define hyperparameter grids for each model in the pipelines dictionary.



```python
param_grid = {
    'RandomForest': {
        'classifier__n_estimators': [100, 200, 300],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 4],
        'classifier__class_weight': ['balanced']
    },
    'GradientBoosting': {
        'classifier__n_estimators': [100, 200, 300],
        'classifier__learning_rate': [0.01, 0.05, 0.1],
        'classifier__max_depth': [3, 4, 5],
        'classifier__subsample': [0.8, 0.9, 1.0],
        'classifier__min_samples_leaf': [1, 5, 10]
    },
    'LogisticRegression': {
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__penalty': ['l1', 'l2']
    }
}
```

## Perform hyperparameter tuning with cross-validation

### Subtask:
Use GridSearchCV or RandomizedSearchCV with cross-validation to find the best hyperparameters for each model, optimizing for a relevant metric for imbalanced data (e.g., F1-score, Precision, or Recall).


**Reasoning**:
Define the F1 scorer and iterate through the pipelines to perform GridSearchCV for each model using the defined parameter grids and the F1 scorer.



```python
f1_scorer = make_scorer(f1_score)

best_estimators = {}

for name, pipeline in pipelines.items():
    print(f"Performing GridSearchCV for {name}")
    grid_search = GridSearchCV(pipeline, param_grid[name], cv=KFold(n_splits=5, shuffle=True, random_state=42), scoring=f1_scorer, n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)
    best_estimators[name] = grid_search.best_estimator_
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best F1 score for {name}: {grid_search.best_score_}")
```

**Reasoning**:
Continue with the next steps of the subtask, which involves performing GridSearchCV for the remaining models and storing their best estimators.



```python
for name, pipeline in pipelines.items():
    if name not in best_estimators: # Skip models already processed in case of interruption
        print(f"Performing GridSearchCV for {name}")
        grid_search = GridSearchCV(pipeline, param_grid[name], cv=KFold(n_splits=5, shuffle=True, random_state=42), scoring=f1_scorer, n_jobs=-1, verbose=1)
        grid_search.fit(X_train, y_train)
        best_estimators[name] = grid_search.best_estimator_
        print(f"Best parameters for {name}: {grid_search.best_params_}")
        print(f"Best F1 score for {name}: {grid_search.best_score_}")
```
