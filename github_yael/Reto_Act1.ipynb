```python
# Assuming 'base' is your DataFrame
quality_counts = base['customer_age'].value_counts()

# Using Matplotlib to create a count plot
plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='green')
plt.title('Count Plot of Customer Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = base.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
  plt.subplot(len(numerical_columns), 2, idx)
  sns.histplot(base[feature], kde=True)
  plt.title(f"{feature} | Skewness: {round(base[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
# Assuming 'base' is your DataFrame
plt.figure(figsize=(10, 8))

# Using Seaborn to create a stripplot
sns.stripplot(x="customer_age", y="income", data=base, hue="customer_age", legend=False, palette='viridis', alpha=0.5) # Using stripplot and addressing future warning

plt.title('Stripplot for customer age and income')
plt.xlabel('Customer Age')
plt.ylabel('Income')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
base['income'] = base['income'].astype(str) # Convert 'quality' to categorical

plt.figure(figsize=(10, 8))

# Using Seaborn to create a violin plot
sns.violinplot(x="income", y="customer_age", data=base, palette="Set2", hue="income", alpha=0.7)

plt.title('Violin Plot for Income and Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Income')
plt.show()
```

```python
#plotting box plot between alcohol and quality
sns.boxplot(x='customer_age', y='income', data=base)
```

```python
# Assuming 'base' is your DataFrame
plt.figure(figsize=(15, 10))

# Select only numerical columns for correlation calculation
numerical_base = base.select_dtypes(include=[np.number])

# Using Seaborn to create a heatmap
sns.heatmap(numerical_base.corr(), annot=True, fmt='.2f', cmap='Pastel2',
linewidths=2)

plt.title('Correlation Heatmap (Numerical Variables)')
plt.show()
```

```python
# Identify categorical columns
categorical_columns = base.select_dtypes(include=["object"]).columns

# Display frequency counts for each categorical feature
for feature in categorical_columns:
  print(f"Frequency counts for '{feature}':")
  display(base[feature].value_counts())
  print("\n" + "="*50 + "\n") # Separator for better readability
```
